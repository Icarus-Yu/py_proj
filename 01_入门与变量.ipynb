### python中变量是标签不是盒子，变量只是对象的引用，
 类型属于对象，不是变量
 Python 变量不是存“值”的容器，而是指向对象的一个名字引用。
 这个对象一直在容器内，如果没有通过标签对它进行引用那么将被垃圾回收

```python
# Python的int没有大小限制！
x = 10
y = 999999999999999999999999999  # C++会溢出，Python不会
z = 0b1010      # 二进制：10
w = 0o12        # 八进制：10
h = 0xA         # 十六进制：10
# 浮点数
pi = 3.14
e = 2.718
big = 1.5e10    # 科学计数法：1.5 × 10^10
```

Python没有 short、long、long long 的区分
没有 float 和 double 的区分，只有 float
int可以无限大（只受内存限制）

```python
# C++没有原生复数，Python有！
c = 3 + 4j      # j表示虚部
print(c.real)   # 3.0
print(c.imag)   # 4.0
```

```python
is_student = True   # 注意：首字母大写
is_working = False

# 布尔运算
result = True and False    # False
result = True or False     # True
result = not True          # False
print(result)
```

使用关键字来进行运算而不是符号

```python
# 字符串
# 单引号、双引号都可以
name = "Icarus"
name = 'Icarus'     # 完全等价

# 三引号：多行字符串
text = """这是第一行
这是第二行
这是第三行"""

# 字符串是不可变的（immutable）
s = "hello"
# s[0] = 'H'  # ❌ 错误！不能修改
s = "Hello"   # ✅ 可以重新赋值
```

```python
# 字符串拼接
greeting = "Hello" + " " + "Icarus"

# 字符串重复
laugh = "ha" * 3    # "hahaha"

# 字符串索引（从0开始，支持负索引）
#负数索引指的是倒数
s = "Python"
print(s[0])     # 'P'
print(s[-1])    # 'n' (最后一个字符)
print(s[-2])    # 'o' (倒数第二个)
```

```python
x = None    # 表示"空"或"无值"

if x is None:
    print("x是空的")
#用is来表示是否等价 
```

```python
#类型查看
x = 10
print(type(x))      # <class 'int'>

y = 3.14
print(type(y))      # <class 'float'>

name = "Icarus"
print(type(name))   # <class 'str'>

#类型转换
# 数字转换
x = int(3.14)       # 3
y = float(10)       # 10.0
z = str(123)        # "123"

# 字符串转数字
num = int("42")     # 42
pi = float("3.14")  # 3.14

# 布尔转换
bool(0)             # False
bool(1)             # True
bool("")            # False (空字符串)
bool("hello")       # True
bool(None)          # False
```
